# Comfort & Stigma Analysis

## *Comfort Discussing Mental Health Issues with Coworkers*

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Replace with your actual column name from the dataset
col_name = "Would you feel comfortable discussing a mental health issue with your coworkers?"

# Dark background
plt.style.use("dark_background")
plt.figure(figsize=(6,4))

# Plot percentage instead of counts (normalize=True)
sns.barplot(
    x=df[col_name].value_counts(normalize=True).values * 100,
    y=df[col_name].value_counts(normalize=True).index,
    palette="rocket"
)

# Add percentage labels
for i, (val, label) in enumerate(zip(
    df[col_name].value_counts(normalize=True).values * 100,
    df[col_name].value_counts(normalize=True).index
)):
    plt.text(val + 1, i, f"{val:.1f}%", color="white", va="center")

plt.title("Comfort Discussing Mental Health Issues with Coworkers", fontsize=14, weight="bold")
plt.xlabel("Percentage")
plt.ylabel("Response")

save_chart("comfort_discussing_mental_health_issues_with_coworkers.png")
plt.show()
```

1. The largest group (35.9%) responded “Maybe”, showing that many employees are uncertain about whether coworkers are safe to talk to.
2. Around 25% feel comfortable (Yes), but this is almost the same as those who responded “No” (24.1%) — meaning that trust is divided.
3. A smaller portion (15%) selected “Unknown”, possibly indicating lack of experience or clarity about how coworkers might react.

This suggests that while some workplaces have open peer support, a large portion of employees remain hesitant due to stigma or fear of being judged.

## *Employer Importance: Physical vs. Mental Health*

```python
# Column names
col_phys = "Overall, how much importance does your employer place on physical health?"
col_ment = "Overall, how much importance does your employer place on mental health?"

# Count frequencies (as percentages) for 0–10 scale
phys_counts = df[col_phys].value_counts(normalize=True).sort_index() * 100
ment_counts = df[col_ment].value_counts(normalize=True).sort_index() * 100

# Ensure both cover all 0–10 values
index = range(11)
phys_counts = phys_counts.reindex(index, fill_value=0)
ment_counts = ment_counts.reindex(index, fill_value=0)

# Plot side-by-side bar chart
bar_width = 0.4
plt.figure(figsize=(10,5))
plt.bar([i - bar_width/2 for i in index], phys_counts, width=bar_width, label="Physical Health", alpha=0.8, color="#66c2a5")
plt.bar([i + bar_width/2 for i in index], ment_counts, width=bar_width, label="Mental Health", alpha=0.8, color="#fc8d62")

# Labels and style
plt.title("Employer Importance: Physical vs. Mental Health", fontsize=14, weight='bold')
plt.xlabel("Importance Rating (0–10)")
plt.ylabel("Percentage of Responses")
plt.xticks(index)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)

save_chart("employer_importance_physical_vs_mental_health.png")
plt.show()
```

1. 🟢 Physical health is prioritized more – Most ratings for physical health cluster around 7–8, showing strong emphasis.
2. 🟠 Mental health lags behind – Ratings for mental health peak at 5, suggesting it’s seen as moderately important, but not as strongly supported.
3. Clear perception gap – At higher importance levels (8–10), physical health dominates, while mental health receives fewer high scores.
4. Mixed views on mental health – Responses are more scattered, showing inconsistency in how employers approach mental health compared to physical health.

# Historical Comparison Analysis

## *Employer providing mental health benefits? (2018-2023)*

```python
col = "Does your employer provide mental health benefits as part of healthcare coverage?"
year_col = "year"

# Ensure year column is integer (handles both strings & numbers)
df[year_col] = df[year_col].astype(str).str.extract(r'(\d{4})').astype(int)

# Count responses per year
yearly_trend = df.groupby([year_col, col]).size().unstack(fill_value=0)

# --- Plot ---
plt.figure(figsize=(10,6))

for response in yearly_trend.columns:
    plt.plot(yearly_trend.index, yearly_trend[response], marker="o", label=response)

plt.title(col, fontsize=12, weight="bold")
plt.xlabel("Year")
plt.ylabel("Number of Responses")
plt.xticks(yearly_trend.index)
plt.legend(title="Response", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()

save_chart("employer_provides_mental_health_benefits.png")
plt.show()
```

1. Strong Majority Said “Yes” (2018–2019) - *In 2018 and 2019, the highest number of respondents confirmed that their employer did provide mental health benefits, showing strong employer support in those years.*

2. Gradual Decline After 2019 - *From 2020 onwards, the number of “Yes” responses dropped steadily. This decline could reflect either fewer employers offering benefits or shifting employee awareness of coverage.*

3. Uncertainty (“I don’t know” & “Unknown”) Remains Noticeable - *Across all years, a significant portion of employees were uncertain about their coverage. This highlights a communication gap between employers and employees about available benefits.*

4. Minimal but Stable “Not eligible for coverage / NA” Group - *Only a small number of respondents consistently indicated that they weren’t eligible for coverage. This remained relatively flat across years.*

5. 2023 Data Limitation ⚠️ - *The sharp drop in all categories in 2023 is not a real trend, but instead due to a very small sample size.*

A steady increase in “Yes” indicates that more organizations began wellness campaigns and formal discussions, likely influenced by the COVID-19 pandemic.

3. Consistent “Unknown” responses : 
Across all years, a portion of employees remained unsure, reflecting communication gaps in how employers conveyed mental health initiatives.

4. 2023 → Very small dataset : 
The drop in all categories is due to low response count, not an actual decline in employer-led mental health discussions.

## *Does tech industry support employees with mental health issues?*

```python
df['year'] = df['year'].astype(str).str.extract(r'(\d{4})')

# Get counts per year and rating
heatmap_data = df.groupby(['year', 'Overall, how well do you think the tech industry supports employees with mental health issues?']).size().unstack(fill_value=0)

# Convert to % per year
heatmap_data = heatmap_data.div(heatmap_data.sum(axis=1), axis=0) * 100

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(heatmap_data.T, annot=True, cmap="YlGnBu", cbar_kws={'label': '% of Responses'}, fmt=".1f")

plt.title("Overall, how well do you think the tech industry supports employees with mental health issues?")
plt.xlabel("Year")
plt.ylabel("Rating (1=Poor, 5=Excellent)")

save_chart("does_tech_industry_support_employees_with_mental_health_issues.png")
plt.show()
```

1. Neutral perception dominates – Across all years (2018–2022), most respondents rated industry support as 3 (≈37–44%), showing that support is viewed as average rather than strong.
2. Decline in negative ratings (1 & 2) until 2021 – Early years (2018–2019) had higher “poor support” ratings (1 & 2 combined ≈47%), which gradually decreased by 2021–2022, indicating improvement in perception.
3. Gradual growth in positive ratings (4 & 5) – Ratings of 4 rose from ~13–15% (2018–2020) to 21% in 2022, while 5 (excellent), though low, increased from ~1% (2018) to ~5% (2021), suggesting slow but steady progress.
4. 2023 outlier due to small sample – The sharp drop in 4 (0%) and spike in 5 (16.7%) in 2023 are artifacts of a very small dataset, not an actual trend.

# Statistical Analysis

## *Age vs. Comfort Discussing Mental Health at Work*

```python
age_col = "What is your age?"
comfort_col = "Would you feel comfortable discussing a mental health issue with your direct supervisor(s)?"

age_bins = [0, 20, 30, 40, 50, 60, 100]
age_labels = ["<20", "20–29", "30–39", "40–49", "50–59", "60+"]
df["Age Group"] = pd.cut(df[age_col], bins=age_bins, labels=age_labels, right=False)

age_comfort = df.groupby(["Age Group", comfort_col]).size().reset_index(name="Count")

plt.figure(figsize=(12,5))
sns.barplot(
    data=age_comfort,
    x="Age Group",
    y="Count",
    hue=comfort_col,
    palette="magma"
)

plt.title("Age vs. Comfort Discussing Mental Health at Work", fontsize=14, weight="bold")
plt.xlabel("Age Group",fontsize=12)
plt.ylabel("Number of Responses",fontsize=12)
plt.legend(title="Comfort Level")
plt.xticks(rotation=45)
plt.tight_layout()
save_chart("Age_vs_Comfort_Discussing_Mental_Health_at_Work.png")
plt.show()
```

1. Mid-career employees (30–39) are the most open – This group shows the highest “Yes” responses, suggesting they feel more comfortable discussing mental health at work.
2. Younger employees (20–29) are divided – While many answered “Yes,” a large share also chose “No” or “Maybe,” reflecting uncertainty early in their careers.
3. Older employees (40+) show more caution – Openness declines in the 40–49 and 50–59 groups, where “No” and “Maybe” responses are more common.
4. Very low participation among 60+ – Few responses in this age group, mostly “Maybe” or “Unknown,” with very limited “Yes.”
